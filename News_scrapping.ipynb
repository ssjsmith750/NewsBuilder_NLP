```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get('https://www.bbc.com/news/world-middle-east-67855117')

time.sleep(5)

# Use find_elements to get a list of all h1 elements on the page
h1_elements = driver.find_elements(By.XPATH, '//p[@class="ssrcss-1q0x1qg-Paragraph e1jhz7w10"]')

# Print the text content of each h1 element
for h1_element in h1_elements:
    print("Title:", h1_element.text)

#time.sleep(5)

```

```python
news_list =[]
```

```python
for h1_element in h1_elements:
    news_list.append(h1_element.text)
```

```python
df=pd.DataFrame(news_list)
```
